```python
import csv
from csv import DictReader

#找出pm2.5小於10的站點
a = []
b = {}
with open("aqx_p_488.csv",encoding='utf-8',newline='') as csv_file:
  reader:DictReader = DictReader(csv_file)
  aqx_p_488 = list(reader)
  
  for i in aqx_p_488:
    if not i.get('pm2.5'):  
      continue
    else:
     if int( i['pm2.5']) < 10:
      b={
        '地區':i["sitename"], 'pm2.5':i['pm2.5']
      }
      a.append(b)

  print("PM2.5小於10的站點:")
  for i in a:
    print(f"地區:{(i['地區']):10s}\tPM2.5:{(i['pm2.5'])}")
```
