```python
#1家分店30天的營收

import numpy as np
np.random.seed(42)
oneStore = np.random.randint(100, 10000, size=(30,))
#建立累計營收
oneStoreCum = oneStore.cumsum()
oneStoreCum
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('ChineseFont.ttf')
mpl.rc('font',family='ChineseFont')
plt.plot(range(1,31),oneStoreCum,color='red',linestyle='-',marker='')
plt.title("分店營收累計")
plt.xlabel("天")
plt.ylabel("累計營收")
plt.show()
```

```python
#5家分店30天的營收
np.random.seed(42)
fiveStore = np.random.randint(100, 10000, size=(5, 30))
fiveStoreCum = fiveStore.cumsum(axis=1)
fiveStoreCum
```

```python
fontManager.addfont('ChineseFont.ttf')
mpl.rc('font',family='ChineseFont')
plt.plot(range(1,31),fiveStoreCum[0],color='r',linestyle='-',marker='',label="第1家分店")
plt.plot(range(1,31),fiveStoreCum[1],color='y',linestyle='-',marker='',label="第2家分店")
plt.plot(range(1,31),fiveStoreCum[2],color='g',linestyle='-',marker='',label="第3家分店")
plt.plot(range(1,31),fiveStoreCum[3],color='m',linestyle='-',marker='',label="第4家分店")
plt.plot(range(1,31),fiveStoreCum[4],color='c',linestyle='-',marker='',label="第5家分店")
plt.title("每家分店營收累計")
plt.xlabel("天")
plt.ylabel("累計營收")
plt.legend()
plt.show()
```
